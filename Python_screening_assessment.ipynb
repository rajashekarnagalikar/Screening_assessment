### Inheritance and multiple Inheritance

```python
# Example
class car():
    def __init__(self, engine, model):
        self.engine =engine
        self.model = model
    
    def description(self):
        print("This method includes the engine and model description of cars.")
        
class car_cost(car):        
    def cost_description(self):
        print("This method includes the cost description of the cars.")
    
class car_profit(car_cost):
    def car_profit(self):
        print("This method includes the car description of engine and model, its cost as well as cars profit.")

```

```python
details = car("rx100","BMW")
```

```python
details.description()
```

```python
cost = car_cost("rx100","BMW")
```

```python
cost.cost_description()
```

### Abstraction

```python
# Example - Lets try to create a shape or printing a shape

from abc import ABC, abstractmethod
# Here we are creating shape class as a parent class first which takes input parameter as shape name 
class Shape(ABC):
    def __init__(self, shape_name):
        self.shape_name = shape_name
    
    # Here we will be inhereting the class draw with other properties
    @abstractmethod
    def draw(self):
        pass
```

```python
from abc import ABC, abstractclassmethod
# Lets consider circle shape to print 
class Circle(Shape):
    def __init__(self):
        super().__init__("circle")
 
    def draw(self):
        print("Drawing a Circle")
```

```python
# creating a object of that class and fetching the output 
circle = Circle()
circle.draw()
```
